1. Determine tres números positivos cuya suma sea 21, tal que su producto P sea un máximo.




Por Multiplicadores de Lagrange:

Se define:


$$\nabla FO = λ \nabla FR$$

En este problema son establecidas dos funciones:

$$\left\lbrace\begin{array}{c} x+y+z = 21 \\ P = xyz \end{array}\right. $$

$$$$

Para encontrar el grandiente de la funcion objetivo $\nabla FO$ Se deriva en $P$ respecto a cada una de sus variables, donde obtenemos:

$$\left\lbrace\begin{array}{c} P_x = yz \\ P_y = xz \\ P_z = xy \end{array}\right.    ⇒ \nabla P(x,y,z) = (P_x, P_y, P_z)$$

Lo mismo se realiza para la funcion restriccion:

$$\left\lbrace\begin{array}{c} g_x = 1 \\ g_y = 1 \\ g_z = 1 \end{array}\right.    ⇒ \nabla g(x,y,z) = (1, 1, 1)$$


De los anteriores gradientes podemos establecer la siguiente relacion:

$$P_x = λ g_x$$

$$P_y = λ g_y$$

$$P_z = λ g_z$$

Remplazando con los calores antes encontrados, obtenemos:

$$yz = λ (1)$$

$$xz= λ (1)$$

$$yz = λ (1)$$

Por lo que podemos decir que,

$$yz = xz = xy$$

En este punto, se estableceran tantas ecuaciones como combinaciones tenga nuesrta formula general, en este caso, las ecuaciones mencionadas se agrupan en las tres siguientes:

1. $$yz = xz$$
2. $$xz = xy$$
3. $$yz = xy$$

Para $(1)$:

$$yz - xz = 0$$
$$z(y-x) = 0$$
$$z = 0  \lor y = x $$

Para $(2)$:
$$xz = xy$$
$$xz -xy = 0$$
$$x(z-y) = 0$$
$$x = 0 \lor z= y$$


Para $(3)$:
$$yz = xy$$
$$yz-xy=0$$
$$y(z-x) = 0$$
$$y = 0 \lor z = x$$


A partir del desarrollo de las tres anteriores igualdades podemos establecer que

$$x = y = z$$

Y regresando a nuestra funcion restriccion tenemos que:

$$x+x+x = 21$$
$$3x = 21$$
$$x = 7$$
$$\therefore x, y, z = 7 $$

Entonces, $P=7*7*7=343$

Consideremos los valores limite de las variables por un momento, si una de las variables se acerca a $0$ o $21$ (los extremos del intervalo posible dado por la restriccion), entonces el valor de P se volvera 0, ya que estara multiplicando por 0.

Por lo tanto, el valor maximo de $P$ ocurre cuando $x=y=z=7$, y $P=343$. Esto significa que el punto que encontramos es un maximo, como fue requerido.

$$\left\lbrace\begin{array}{c} S_l = ah \\ S_a = lh \\ S_h = la \end{array}\right.    ⇒ \nabla FR(l,a,h) = (S_l, S_a, S_h)$$

$$P_l = λS_l =2a+2h = λ (ah)$$
$$P_a = λS_a=2l + 2h = λ(lh)$$
$$P_h = λS_h=2l +2a = λ(la)$$

$$$$

$$$$
$$\frac{2l+2h}{lh} = λ$$

$$h \neq 0$$

 $$\begin{align*}  & \left( \frac{a+h}{a} - \frac{l+h}{l}  \right)

$$a , h \neq 0$$

$$a(l+h) - h(l+a) = 0$$

Operando:

$$al + ah -hl -ha = 0$$

$$ al -hl  = 0$$

$$ l(a-h) = 0$$

$$a=h$$

$$\frac{2a+2h}{ah} -\frac{2l+2a}{la} = 0$$

$$\frac{2}{a} \begin{align*}  & \left( \frac{a+h}{h} - \frac{l+a}{l}  \right)
\end{align*} = 0$$

De aqui, se establece que

$$a \neq 0$$

Por lo tanto, continuamos con lo que resta de la expresion:


$$\begin{align*}  & \left( \frac{a+h}{h} - \frac{l+a}{l}  \right)

3. Encuentre el punto sobre el plano $x+2y+z=1$ más cercano al origen.

Para abordar este problema, requerimos dos componentes esenciales: una función objetivo y una función de restricción. La función objetivo es la que se optimizará para encontrar un valor máximo o mínimo, mientras que la función de restricción establece las condiciones bajo las cuales se puede lograr dicha optimización. Sin embargo, en el enunciado proporcionado, no se especifica ninguna función objetivo para optimizar.

(Mirese plano bidimensional, y tresdimensional)

En un plano bidimensional, si tenemos una recta en un plano $x,y$ y deseamos encontrar el punto de la recta más cercano al origen, podemos utilizar una circunferencia como herramienta. A medida que incrementamos el radio de la circunferencia, eventualmente llegará un punto en el que intersectará con nuestra recta. En este caso, ese punto de intersección será el más cercano al origen.

Extrapolando este concepto a un plano tridimensional, podemos utilizar una esfera en lugar de una circunferencia. Al incrementar el tamaño de la esfera, llegará un momento en el que esta intersectará con el plano. En este contexto, utilizaremos la función de la esfera como nuestra función objetivo que será optimizada.


La formula de una esfera es:


$$x^2+y^2+ z^2 = r^2$$ Que es nuestra funcion objetivo.

Del anterior enunciado se establecen dos funciones.


$$\left\lbrace\begin{array}{c} x+2y+z = 1 \\ x^2 +y^2 +z^2 = r^2 \end{array}\right. $$

$$\nabla FO = λ \nabla FR$$

$$\left\lbrace\begin{array}{c} FO_x = 2x \\ FO_y =2y \\ FO_z =2z \end{array}\right.    ⇒ \nabla FO(x,y,z) = (FO_x, FO_y, FO_z)$$

Lo mismo se realiza para la funcion restriccion:

$$\left\lbrace\begin{array}{c} FR_x = 1 \\ FR_y = 2 \\ FR_z = 1 \end{array}\right.    ⇒ \nabla FR(x,y,z) = (FR_x , FR_y, FR_z)$$


De los anteriores gradientes podemos establecer la siguiente relacion:

$$FO_x = λFR_x = 2x = λ (1)$$
$$FO_y = λFR_y = 2y = λ(2)$$
$$FO_z = λFR_z = 2z = λ(1)$$

De las anteriores relaciones, despejando $λ$ se obtiene:

$$2x = λ$$

$$y = λ$$

Una lata de refresco se construye con un costado lateral de estaño y una tapa y fondo de aluminio. Dado que el costo es de 1.8 centavos por unidad cuadrada de la tapa, 1 centavo por unidad cuadrada del fondo y 2.3 centavos por unidad cuadrada del costado, determine la función de costo \( C(r, h) \), donde \( r \) es el radio de la lata y \( h \) es su altura.

**Función de costo \( C(r, h) \)**

La lata de refresco tiene una forma cilíndrica, y su superficie se compone de tres partes:

**Superficies y costos:**

1. **Área de la tapa (aluminio):**
   $$
   \text{Área de la tapa} = \pi r^2
   $$
   Costo de la tapa = 1.8 centavos por unidad cuadrada

2. **Área del fondo (aluminio):**
   $$
   \text{Área del fondo} = \pi r^2
   $$
   Costo del fondo = 1 centavo por unidad cuadrada

3. **Área del costado lateral (estaño):**
   $$
   \text{Área del costado} = 2\pi rh
   $$
   Costo del costado = 2.3 centavos por unidad cuadrada

Sumamos estos costos para obtener la función de costo total:

$$
C(r, h) = C_{\text{tapa}} + C_{\text{fondo}} + C_{\text{costado}}
$$

$$
C(r, h) = 1.8 \pi r^2 + 1 \pi r^2 + 4.6 \pi r h
$$

$$
C(r, h) = (1.8 + 1) \pi r^2 + 4.6 \pi r h
$$

$$
C(r, h) = 2.8 \pi r^2 + 4.6 \pi r h
$$

Por lo tanto, la función de costo \( C(r, h) \) en términos del radio \( r \) y la altura \( h \) de la lata es:

$$
C(r, h) = 2.8 \pi r^2 + 4.6 \pi r h
$$

**Ahora optimicemos un poquito**

$$
C(r, h) = 2.8 \pi r^2 + 4.6 \pi r h
$$

Los puntos críticos los encontramos al igualar a cero las derivadas parciales de C :

$$\frac{∂C}{∂r}  = 5.6πr+4.6πh=0 $$

$$\frac{∂C}{∂h}  = 4.6πh=0 $$

Donde procederemos a resolver mediante sustitucion:

$$4.6πr=0 $$
$$r=0$$

$r=0$ en la primera ecuación:

$$ 5.6π(0)+4.6πh=0 $$
$$ 4.6πh=0 $$
$$ h=\frac{0}{4.6π} $$
$$ h=0 $$

**Nuestro punto crítico es**
$$(r,h)=(0,0)$$

Por lo tanto nos damos cuenta que es hora de enfrentarnos al punto silla

Para evitar este punto silla procederemos a incluir una restricción que tenga sentido para nosotros en este caso sera el volumen de la lata puesto que no podemos tener un volumen variable, ¿Para que querriamos un volumen variable ?

$$h=\frac{V}{π*r^2}$$
